```python
#select population column
data['Population']
```

```python
#Clean the population column by replacing commas
my_clean_population = []
for i in data['Population']:
    my_clean_population.append(i.replace(',',''))
my_clean_population
```

```python
#create a new column for clean population
data['clean_pop'] = my_clean_population
data
```

```python
#Smart way of doing the cleaning
data['smart_clean_pop'] = data['Population'].apply(lambda x : x.replace(',',''))
data
```

```python
data['clean_curr'] = data['Exchange Rate Against Dollar'].apply(lambda x : x.split()[0].replace(',',''))
data
```

```python
data['title_curr'] = data['Exchange Rate Against Dollar'].apply(lambda x : x.split()[1])
data
```

```python
data['International Soccer Team Title'] = data['International Soccer Team'].apply(lambda x : ' '.join(x.split()[:-3]))
data['remaining'] = data['International Soccer Team'].apply(lambda x : ' '.join(x.split()[-3:]))
data
```

```python
data.drop(['Population', 'Exchange Rate Against Dollar', 'clean_pop'], axis = 1, inplace = True)
data
```

```python
data.rename({'smart_clean_pop' : "Population", 'clean_curr' : "ExchangeRate", 'title_curr' : 'Currency_Title'}, axis = 1, inplace =True)
data
```
